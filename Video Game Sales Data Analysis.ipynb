# Video Game Sales Data Analysis
The data consists of historical video game sales. 
 Categorical variables includes: rank (in sales), title, platform (system), year of release, genre, publisher, and sales
amount (in millions) for the regions of: North America, Europe, Japan, and Other Countries.

# 1. Loading the data to perform analysis 

```python
#importing required libraries to perform data manipulation and analysis
import pandas as pd
import matplotlib as plt
import seaborn as sns 
from IPython.display import display, HTML
```

```python
#loading the dataset into a pandas dataframe 
data=pd.read_csv("HW5_videogamesales.csv")
data=data.set_index('Rank')
total_rows =data.shape[0]  # gives number of row count
print("The total number of video games:",+total_rows)
top_games=data[0:15]
display(top_games) #to display in tabular format 
```

# 2. Adding new global sales data to original dataset.

```python
#manipulating dataframe to calculate sum of sales in different region and add new columns 'global sales' to dataset
sales_data=data[data.columns[5:]]
global_sales=sales_data.sum(axis=1)
data['Global_Sales']= global_sales
top_games_updated=data[0:15]
display(top_games_updated)
```

# 3. Descriptive statistics of Global Sales.
To perform Descriptive statistics, we calculate mean, median, min, max and standard deviation of global sales. This gives us an overview of the sales data.
Furthermore, we also calculate Percentage of video games that earned at least $5,000,000 in global sales starting in the year
2000 or later.

```python
print("The total number of video games:",+total_rows)
mini=str(data['Global_Sales'].min(axis = 0))
print("The minimum sales is:"+mini)
maxx=data['Global_Sales'].max(axis = 0)
print("The maximum sales is:",+maxx)
average = str(data['Global_Sales'].mean())
print("The mean global sales is:"+average)
med = str(data['Global_Sales'].median())
print("The median global sales is:"+med)
stdev= str(data['Global_Sales'].std())
print("The Standard Deviation for global sales is:"+stdev)

percent=data[(data['Year']>2000.0) & (data['Global_Sales']>5)].shape[0]
print(percent)
percent=(percent/total_rows)*100
print("Percentage of video games that earned at least $5,000,000 in global sales starting in the year 2000 or later is:%",+percent)
# we can also do descriptive statistics by describe function, so I have done it using both ways. 
data['Global_Sales'].describe()
```

# 4. Defining python functions. 

```python
def topgames(data,category=None,value=None,number=10):
    if category!= None and value!=None:
        return data[data[category]==value].sort_values(by='Global_Sales', ascending=False).head(number)
    else:
        return data.sort_values(by='Global_Sales', ascending=False).head(number)
```

# 5. Creating dynamic tables based on user requirements

```python
# calling previously defined function to create tables based on different categories and values
display(topgames(data,number=5))
display(topgames(data,'Platform','Wii',10))
display(topgames(data,'Genre','Puzzle',5))
display(topgames(data,'Year',1990,10))

```

# 6. Correlation Matrix

```python
#to create a correlation matrix of sales region with global sales
data.iloc[:,5:].corr()
```

# Observations and Inference: 
All markets have a positive uphill correlation with global sales. 
Meaning if there is and increase in global sales, there will be an increase in every market.

Japan market is moderately correlated whereas rest markets are strongly related to global market.

Inference:
North America sales is most dominant among other region and therefore global sales region is highly dependent on how the sales are in North America. It is because the video game market in North America is very strong and the population that plays video games is more compared to any other region in the market.

Other sales region comprises of many smaller markets around the world. Each market individually does not affect much of global sales but when aggregated the sales contribute a good amount to the global sales. so companies should try to increase their sales in all the smaller markets and there are more chances of success as the market is niche and they could earn more profits.


# 7. Understanding the relationship between sales during 2000 and 2015.

```python
display(data[data.Year==2000].iloc[:,5:].corr())

display(data[data.Year==2015].iloc[:,5:].corr())
```

# Observations and Inference:

Over the period of 15 years, all markets have become more dependent(correlated) on global market except for Japan. If global sales increases, Japan's sales would increase but not by much as it is weakly correlated to global sales.
During 2000, Japan market is moderately correlated whereas rest markets are strongly related to global market.
During 2015, EU sales have increased, leading to more constribution towards global sales, followed by other region sales. So we can conclude that sellers must have incorporated new policies to boost sales in smaller market so as to expand their global market sales.
Moreover, we can infer that japan's contribution to global sales has decreased over the period of 15 years and therefore the company should implement policies to increase japan's sales.
